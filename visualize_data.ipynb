```python
import matplotlib.pyplot as plt
import numpy as np
from chainercv import transforms
import pathlib
import cv2
```

# visualize data

```python
data_path = pathlib.Path("../data/CelebA/Img/img_align_celeba/").resolve()
save_path = pathlib.Path("../data/CelebA/Img/center_cropped_resize_64").resolve()
if not save_path.exists():
    save_path.mkdir()
```

```python
height, width = 218, 178
dx,dy = 128, 128
is_resize=True
#cnt=0
for path in data_path.glob("*.jpg"):
    #f cnt<10:
    img = cv2.imread(str(path))
    #plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #plt.show()
    cropped_img = img[height//2 - dx // 2:height//2 + dx // 2,
                          width//2 - dy // 2:width//2 + dy // 2]
    if is_resize:
        cropped_img= cv2.resize(cropped_img, dsize=(64, 64),
                                interpolation=cv2.INTER_LANCZOS4)
    #plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    #plt.show()

    # save
    cv2.imwrite(
        str(save_path / path.name), cropped_img)
```

```python
save_path = pathlib.Path("../data/CelebA/Img/center_cropped_128").resolve()
if not save_path.exists():
    save_path.mkdir()

height, width = 218, 178
dx,dy = 128, 128
is_resize=False
#cnt=0
for path in data_path.glob("*.jpg"):
    #f cnt<10:
    img = cv2.imread(str(path))
    #plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    #plt.show()
    cropped_img = img[height//2 - dx // 2:height//2 + dx // 2,
                          width//2 - dy // 2:width//2 + dy // 2]
    if is_resize:
        cropped_img= cv2.resize(cropped_img, dsize=(64, 64),
                                interpolation=cv2.INTER_LANCZOS4)
    #plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
    #plt.show()

    # save
    cv2.imwrite(
        str(save_path / path.name), cropped_img)
```

```python
cnt=0
for path in save_path.glob("*jpg"):
    img = cv2.imread(str(path))
    fig, ax = plt.subplots(1,1, figsize=(2,2))
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    fig.show()
    cnt+=1
    if cnt>10:
        break
```
